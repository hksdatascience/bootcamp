```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
my_page = requests.get(
    "https://www.businessinsider.com/best-suburbs-to-live-in-america-ranked-2018-9")
print(my_page.status_code) #200


```

```python
suburb_soup = BeautifulSoup(my_page.content, 'html.parser')
```

```python
#print (suburb_soup.body.prettify()[:])
```

```python
suburb=suburb_soup.find_all(class_="slide-title-text")
suburb_text = suburb_soup.find_all("p")
```

```python
name = []
rank = []
for burb in suburb:
    rank.append(int(burb.text[:2].replace(".","")))
    name.append(burb.text[3:])
```

```python
y = 0
population=[]
home_val = []
hh_income = []
crime_grade = []

for x in suburb_text[8:]:
    if "Population" in x.text:
        population.append(x.text[12:])
    if "Median home value" in x.text:
        home_val.append(x.text[20:])
    if "Median household income" in x.text:
        hh_income.append(x.text[26:])
    if "Crime & safety grade" in x.text:
        crime_grade.append(x.text[22:])
    
```

```python
df = pd.DataFrame({"City":name,
                  'Population':population,
                  "Median Home Value":home_val,
                  "Median Household Income":hh_income,
                  "Crime & Safety Grade":crime_grade})

df.index = rank
df.index.name = 'rank'
```

```python
df.sort_index(ascending=True)
```
